```python
#Dealing with null values
def NullRemoval(a):
    a.MSZoning.fillna(a.MSZoning.mode()[0], inplace = True)
    a.LotFrontage = a.groupby('Neighborhood')['LotFrontage'].transform(lambda x: x.fillna(x.mean()))
    a.Alley.fillna('None', inplace = True)
    a.Utilities.fillna('None',inplace = True)
    a.Exterior1st.fillna(a.Exterior1st.mode()[0], inplace = True)
    a.Exterior2nd.fillna(a.Exterior2nd.mode()[0], inplace = True)
    a.MasVnrType.fillna('None', inplace = True)
    a.MasVnrArea.fillna(0, inplace = True)
    a.BsmtQual.fillna('None', inplace = True)
    a.BsmtCond.fillna('None', inplace = True)
    a.BsmtExposure.fillna('None', inplace = True)
    a.BsmtFinType1.fillna('None', inplace = True)
    a.BsmtFinType2.fillna('None', inplace = True)
    a.BsmtFinSF1.fillna(0, inplace = True)
    a.BsmtFinSF2.fillna(0, inplace = True)
    a.BsmtUnfSF.fillna(0, inplace = True)
    a.TotalBsmtSF.fillna(0, inplace = True)
    a.Electrical.fillna(a.Electrical.mode()[0], inplace = True)
    a.BsmtFullBath.fillna(0, inplace = True)
    a.BsmtHalfBath.fillna(0, inplace = True)
    a.KitchenQual.fillna(a.KitchenQual.mode()[0], inplace = True)
    a.Functional.fillna('Typ', inplace = True)
    a.FireplaceQu.fillna('None', inplace = True)
    a.GarageType.fillna('None', inplace = True)
    a.GarageCars.fillna(0, inplace = True)
    a.GarageArea.fillna(0, inplace = True)
    a.GarageQual.fillna('None', inplace = True)
    a.GarageCond.fillna('None', inplace = True)
    a.GarageFinish.fillna('None', inplace = True)
    a.PoolQC.fillna('None', inplace = True)
    a.Fence.fillna('None', inplace = True)
    a.MiscFeature.fillna('None', inplace = True)
    a.SaleType.fillna(a.SaleType.mode()[0], inplace = True)
    a.GarageYrBlt.fillna(0, inplace = True)
    
    return a
```

```python
test = NullRemoval(test)
train = NullRemoval(train)
```

```python
def cat_to_ord(a):
    a.replace({
        #'Street' : {'Grvl' : 1, 'Pave' : 2 },
        #'Alley' : {'None' : 0, 'Grvl' : 1, 'Pave' : 2 },
        #'LotShape' : {'IR3' : 1, 'IR2' : 2 , 'IR1' : 3 , 'Reg' : 4 },
        #'Utilities' : {'None' : 0, 'NoSeWa' : 1, 'AllPub' : 2},
        #'LandSlope' : {'Gtl' : 1, 'Mod': 2, 'Sev': 3},
        #'ExterQual' : {'Fa': 1, 'TA': 2, 'Gd': 3, 'Ex': 4},
        #'ExterCond' : {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
        #'BsmtExposure' : {'None': 0, 'No': 1, 'Mn': 2, 'Av': 3, 'Gd': '4'},
        #'BsmtFinType1' : {'None': 0, 'Rec': 1 ,'BLQ': 1 ,'LwQ': 1 ,'ALQ': 1 ,'Unf': 1,'GLQ': 2},
        #'BsmtFinType2' : {'None': 0, 'Rec': 1 ,'BLQ': 1 ,'LwQ': 1 ,'ALQ': 1 ,'Unf': 1,'GLQ': 2},
        #'HeatingQC' : {'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
        #'CentralAir' : {'N': 0, 'Y': 1},
        #'KitchenQual' : {'Fa': 1, 'TA': 2, 'Gd': 3, 'Ex': 4},
        #'FireplaceQu' : {'None': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
        #'GarageFinish' : {'None': 0, 'Unf' : 1, 'RFn': 2, 'Fin': 3},
        #'GarageQual' : {'None': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
        #'GarageCond' : {'None': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
        #'PavedDrive' : {'N': 0, 'P': 1, 'Y': 2},
        #'PoolQC' : {'None': 0, 'Po': 1, 'Fa': 2, 'TA': 3, 'Gd': 4, 'Ex': 5},
        'MSSubClass' :  {20 : "SC20", 30 : "SC30", 40 : "SC40", 45 : "SC45", 
                         50 : "SC50", 60 : "SC60", 70 : "SC70", 75 : "SC75", 
                         80 : "SC80", 85 : "SC85", 90 : "SC90", 120 : "SC120", 
                         150 : "SC150", 160 : "SC160", 180 : "SC180", 190 : "SC190"},
        'MoSold' : {1 : 'Jan' , 2 : 'Feb' , 3 : 'Mar' , 4 : 'Apr' , 5 : 'May' , 6 : 'Jun' , 
                    7 : 'Jul' , 8 : 'Aug' , 9 : 'Sep' , 10 : 'Oct' , 11 : 'Nov' , 12 : 'Dec'}
    
    }, inplace = True)
    return a
```
